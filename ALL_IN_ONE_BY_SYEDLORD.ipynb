```python
#@markdown <h1> <center> <strong> Movies & Series Uploader with Auto Renaming and Metadata üìΩÔ∏èüì∫üìù
import os
import pickle
import subprocess
import requests
import re
import tempfile
from google.auth.transport.requests import Request
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from langdetect import detect
from tqdm.notebook import tqdm
from googleapiclient.errors import HttpError

# Track YouTube API quota usage
quota_usage = {
    "videos.insert": 0,
    "captions.insert": 0,
    "videos.update": 0,
    "thumbnails.set": 0,
    "playlists.insert": 0,
    "playlistItems.insert": 0
}

current_credential_index = 0

# ============================
# User Input: Video Path, Language Selection & Subtitle Option
# ============================
video_path = ""  #@param {type:"string"}
selected_audio_language = "hin"  #@param ["hin", "eng", "fra", "spa"]
subtitle_option = "Upload English Subtitles"  #@param ["Upload All Subtitles", "Upload English Subtitles", "Extract Subtitles", "Extract English Subtitles", "No Subtitles"]
playlist_option = "None"  #@param ["None", "Create New Playlist", "Save to Existing Playlist"]

# Single TMDB link that will be processed accordingly
tmdb_link = ""  #@param {type:"string"}

season_number = ""  #@param {type:"string"}
existing_playlist_id = ""  #@param {type:"string"}
auto_rename_format = True  #@param {type:"boolean"}
fetch_metadata = True  #@param {type:"boolean"}
tmdb_api_key = "c5668e874013b9df2a50fba06875a331"  #@param {type:"string"}

# Ensure correct directory handling
if os.path.isdir(video_path):
    video_dir = video_path
    video_files = [os.path.join(video_dir, f) for f in os.listdir(video_dir) if f.endswith(('.mp4', '.mkv'))]
elif os.path.isfile(video_path):
    video_dir = os.path.dirname(video_path)
    video_files = [video_path]
else:
    raise FileNotFoundError("Invalid path! Provide a valid video file or directory.")

# ============================
# Authentication & API Setup
# ============================
def load_valid_credentials(start_index=0):
    global creds, youtube, current_credential_index

    max_attempts = 10  # credentials0.json to credentials9.json

    for attempt in range(start_index, max_attempts):
        cred_path = f"/content/drive/MyDrive/Credentials/credentials{attempt}.json"

        if not os.path.exists(cred_path):
            print(f"‚ùå credentials{attempt}.json not found, trying next...")
            continue

        print(f"üîê Trying credentials{attempt}.json...")

        token_path = "token.pickle"
        if os.path.exists(token_path):
            os.remove(token_path)

        SCOPES = ["https://www.googleapis.com/auth/youtube.force-ssl"]
        flow = InstalledAppFlow.from_client_secrets_file(cred_path, SCOPES)
        creds = flow.run_console()

        with open(token_path, "wb") as token:
            pickle.dump(creds, token)

        try:
            youtube = build("youtube", "v3", credentials=creds)
            youtube.channels().list(part="id", mine=True).execute()
            current_credential_index = attempt  # ‚úÖ Track the one we're using
            print(f"‚úÖ credentials{attempt}.json is valid and has quota.")
            print(f"üéØ Using credentials{attempt}.json")
            return
        except Exception as e:
            print(f"‚ö†Ô∏è credentials{attempt}.json failed: {e}")
            continue

    raise RuntimeError("‚ùå No credential with remaining quota found.")

# Create a temporary directory for processed files
temp_dir = tempfile.mkdtemp()
print(f"Created temporary directory: {temp_dir}")
# Load YouTube credentials before any upload or metadata work
load_valid_credentials(start_index=current_credential_index + 1)


# ============================
# Metadata Fetching Functions
# ============================

def parse_tmdb_movie_link(link):
    """Extracts the movie ID from a TMDb link."""
    parts = link.strip().split('/')
    if "movie" in parts:
        movie_index = parts.index("movie")
        if len(parts) > movie_index + 1:
            return parts[movie_index + 1].split('-')[0]
    return None

def parse_tmdb_tv_link(link):
    """Extracts the TV show ID and season number from a TMDb link."""
    parts = link.strip().split('/')
    if "tv" in parts:
        tv_index = parts.index("tv")
        if len(parts) > tv_index + 1:
            tv_id = parts[tv_index + 1].split('-')[0]

            # Check if season is in URL
            if "season" in parts:
                season_index = parts.index("season")
                if len(parts) > season_index + 1:
                    season_number_from_url = parts[season_index + 1]
                    return {"id": tv_id, "season_number": season_number_from_url}

            # If season is not in URL but provided separately
            return {"id": tv_id, "season_number": season_number}
    return None

def fetch_movie_metadata(movie_id, api_key):
    """Fetches movie metadata from TMDb."""
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}&language=en-US"
    response = requests.get(url)

    if response.status_code == 200:
        movie_data = response.json()
        return {
            "title": movie_data.get("title", "Unknown Title"),
            "description": movie_data.get("overview", "No description available."),
            "thumbnail": f"https://image.tmdb.org/t/p/w500{movie_data.get('poster_path')}" if movie_data.get("poster_path") else None
        }
    return None

def fetch_episodes(tv_id, season_number, api_key):
    """Fetches episode metadata for a given season."""
    url = f"https://api.themoviedb.org/3/tv/{tv_id}/season/{season_number}?api_key={api_key}&language=en-US"
    response = requests.get(url)

    if response.status_code == 200:
        season_data = response.json()
        episodes = {
            str(ep["episode_number"]): {
                "title": f"{ep['episode_number']}. {ep['name']}",
                "description": ep.get("overview", "No description available."),
                "thumbnail": f"https://image.tmdb.org/t/p/w500{ep.get('still_path')}" if ep.get("still_path") else None
            }
            for ep in season_data.get("episodes", [])
        }
        return episodes
    return None

def fetch_tv_show_info(tv_id, api_key):
    """Fetches TV show information."""
    url = f"https://api.themoviedb.org/3/tv/{tv_id}?api_key={api_key}&language=en-US"
    response = requests.get(url)

    if response.status_code == 200:
        show_data = response.json()
        return {
            "name": show_data.get("name", "Unknown Show"),
            "description": show_data.get("overview", "No description available.")
        }
    return None

# ============================
# Function to Extract TMDB Info
# ============================
def get_tmdb_info(tmdb_link, season_number):
    if not tmdb_link or "themoviedb.org" not in tmdb_link:
        print("‚ùå Invalid TMDB link")
        return None, None, None

    # Ensure the link doesn't end with /seasons or /season/X already
    if tmdb_link.endswith("/seasons"):
        tmdb_link = tmdb_link[:-8]  # Remove "/seasons" from the end

    if "/season/" in tmdb_link:
        tmdb_link = tmdb_link.split("/season/")[0]  # Remove "/season/X" from the end

    # Extract show ID from the URL
    show_id_match = re.search(r'tv/(\d+)-?', tmdb_link)

    if not show_id_match:
        print("‚ùå Unable to extract show ID")
        return "New Playlist", "No description available", None

    show_id = show_id_match.group(1)

    # First get the show details to get the show name
    show_url = f"https://api.themoviedb.org/3/tv/{show_id}?api_key={tmdb_api_key}&language=en-US"
    try:
        show_response = requests.get(show_url)
        show_response.raise_for_status()
        show_data = show_response.json()
        show_title = show_data.get("name", "Unknown Show")
    except requests.exceptions.RequestException as e:
        print(f"‚ùå Error fetching show data: {e}")
        show_title = "Unknown Show"

    # Now get the season details
    season_url = f"https://api.themoviedb.org/3/tv/{show_id}/season/{season_number}?api_key={tmdb_api_key}&language=en-US"

    try:
        season_response = requests.get(season_url)
        season_response.raise_for_status()
        season_data = season_response.json()

        # Extract description
        season_name = season_data.get("name", f"Season {season_number}")
        description = season_data.get("overview", "No description available").strip()

        # Construct playlist title using show name and season number
        playlist_title = f"{show_title}: {season_name}"

        return playlist_title, description, None
    except requests.exceptions.RequestException as e:
        print(f"‚ùå Error fetching season data: {e}")
        return f"{show_title}: Season {season_number}", "No description available", None

# ============================
# Function to Create or Get Playlist
# ============================
def handle_playlist(playlist_option, video_ids):
    global youtube  # use latest credentials

    if playlist_option == "None" or not video_ids:
        return

    playlist_id = None

    # Create or find playlist
    while True:
        try:
            if playlist_option == "Create New Playlist":
                if not tmdb_link:
                    print("‚ùå No TMDB link provided. Skipping playlist creation.")
                    return

                playlist_tmdb_link = f"{tmdb_link}/seasons" if not tmdb_link.endswith("/seasons") else tmdb_link
                playlist_title, playlist_description, _ = get_tmdb_info(playlist_tmdb_link, season_number)

                playlist_insert_response = youtube.playlists().insert(
                    part="snippet,status",
                    body={
                        "snippet": {
                            "title": playlist_title,
                            "description": playlist_description
                        },
                        "status": {
                            "privacyStatus": "private"
                        }
                    }
                ).execute()

                quota_usage["playlists.insert"] += 50

                playlist_id = playlist_insert_response["id"]
                print(f"‚úÖ Created new playlist: {playlist_title} (ID: {playlist_id})")

            elif playlist_option == "Save to Existing Playlist":
                if not existing_playlist_id:
                    print("‚ùå No existing playlist ID provided. Skipping playlist addition.")
                    return

                if "list=" in existing_playlist_id:
                    playlist_id = existing_playlist_id.split("list=")[1].split("&")[0]
                else:
                    playlist_id = existing_playlist_id

                print(f"‚úÖ Using existing playlist ID: {playlist_id}")
            break  # ‚úÖ playlist created or found
        except HttpError as e:
            if e.resp.status == 403 and "quota" in str(e).lower():
                print("‚õî Quota exceeded during playlist creation. Switching credentials...")
                load_valid_credentials(start_index=current_credential_index + 1)
                continue
            else:
                raise

    # Add videos to playlist
    playlist_pbar = tqdm(video_ids, desc="Adding to playlist", colour="green")
    for video_id in playlist_pbar:
        while True:
            try:
                youtube.playlistItems().insert(
                    part="snippet",
                    body={
                        "snippet": {
                            "playlistId": playlist_id,
                            "resourceId": {
                                "kind": "youtube#video",
                                "videoId": video_id
                            }
                        }
                    }
                ).execute()
                quota_usage["playlistItems.insert"] += 50
                playlist_pbar.set_description(f"Added video {video_id}")
                break
            except HttpError as e:
                if e.resp.status == 403 and "quota" in str(e).lower():
                    print(f"‚õî Quota exceeded during playlist insert for video {video_id}. Switching credentials...")
                    load_valid_credentials(start_index=current_credential_index + 1)
                    continue
                else:
                    playlist_pbar.set_description(f"Error: {str(e)[:20]}...")
                    break
    print(f"‚úÖ Playlist operations complete")

# ============================
# ENHANCED Function to Extract Subtitles to Temp Directory
# ============================
def extract_subtitles(video_file, english_only=False):
    # Create a subtitles folder inside the temp directory
    subtitle_folder = os.path.join(temp_dir, f"subtitles_{os.path.basename(video_file)}")
    os.makedirs(subtitle_folder, exist_ok=True)
    extracted_files = []

    language_map = {
        "eng": "English",
        "hin": "Hindi",
        "spa": "Spanish",
        "fra": "French",
        "ger": "German",
        "ita": "Italian",
        "jpn": "Japanese",
        "kor": "Korean",
        "chi": "Chinese",
        "rus": "Russian",
        "por": "Portuguese",
        "ara": "Arabic"
    }

    # Track language counts to handle duplicates
    language_counts = {}

    if video_file.endswith(".mkv"):
        try:
            print(f"Analyzing subtitle tracks in {os.path.basename(video_file)}...")
            track_info = subprocess.check_output(["mkvmerge", "-i", video_file]).decode()
            subtitle_tracks = [line for line in track_info.split("\n") if "subtitles" in line.lower()]

            if not subtitle_tracks:
                print("‚ö†Ô∏è No subtitle tracks found")
                return extracted_files

            print(f"Found {len(subtitle_tracks)} subtitle track(s)")

            # Create progress bar for subtitle extraction
            subtitle_pbar = tqdm(subtitle_tracks, desc="Extracting subtitles", colour="blue")

            for track in subtitle_pbar:
                track_id = track.split(':')[0].split()[-1]
                subtitle_ext = "ass" if "ass" in track.lower() else "srt"

                # Extract language code if present
                lang_match = re.search(r"\[language:(\w+)\]", track.lower())
                lang_code = lang_match.group(1) if lang_match else "unk"
                lang_name = language_map.get(lang_code, "Unknown")

                # Check if we've seen this language before to handle duplicates
                if lang_name in language_counts:
                    language_counts[lang_name] += 1
                    subtitle_name = f"{lang_name} {language_counts[lang_name]}.{subtitle_ext}"
                else:
                    language_counts[lang_name] = 1
                    subtitle_name = f"{lang_name}.{subtitle_ext}"

                subtitle_pbar.set_description(f"Extracting {lang_name}")
                output_file = os.path.join(subtitle_folder, subtitle_name)

                extract_cmd = ["mkvextract", "tracks", video_file, f"{track_id}:{output_file}"]
                result = subprocess.run(extract_cmd, check=False, capture_output=True, text=True)

                if result.returncode == 0 and os.path.exists(output_file):
                    # Check if language info is present in the track
                    has_language_info = "[language:" in track.lower()

                    # Filter logic based on english_only parameter
                    if english_only:
                        # Keep the file if it's English OR if it has no language information
                        if "[language:eng]" in track.lower() or not has_language_info:
                            extracted_files.append(output_file)
                            subtitle_pbar.set_description(f"‚úÖ Extracted {lang_name}")
                        else:
                            # Not English, remove the file
                            os.remove(output_file)
                            subtitle_pbar.set_description(f"‚ùå Skipped {lang_name} (not English)")
                    else:
                        # Extract all subtitles
                        extracted_files.append(output_file)
                        subtitle_pbar.set_description(f"‚úÖ Extracted {lang_name}")
                else:
                    subtitle_pbar.set_description(f"‚ùå Failed to extract track {track_id}")

        except subprocess.CalledProcessError as e:
            print(f"‚ùå MKVToolNix error: {e}")

    return extracted_files

# ============================
# Function to Determine Video Name for Upload
# ============================
def get_upload_name(video_file, index):
    """
    Determines the name to use for YouTube upload.
    Returns the episode number for series, otherwise returns original name.

    @param video_file: The path to the video file
    @param index: The index of the file in the list (for simple numbering)
    @return: A name for YouTube upload
    """
    if not auto_rename_format:
        return os.path.splitext(os.path.basename(video_file))[0]

    # Check if the file has SXXEXX format (series episode)
    filename = os.path.basename(video_file)
    episode_match = re.search(r"S\d{2}E(\d{2})", filename, re.IGNORECASE)

    if episode_match:
        # Extract episode number and use it as the name
        episode_number = episode_match.group(1)
        return str(int(episode_number))  # Convert to int to remove leading zeros
    else:
        # If not a series episode, use the index+1 as the upload name
        return str(index + 1)

# ============================
# Function to Check if Video is a Series Episode
# ============================
def is_series_episode(video_file):
    """Check if the video file appears to be a TV series episode based on filename."""
    filename = os.path.basename(video_file)
    episode_match = re.search(r"S\d{2}E\d{2}", filename, re.IGNORECASE)
    return bool(episode_match)

# ============================
# Function to Update Video Metadata
# ============================
def update_video_metadata(video_id, title, description, thumbnail_url=None):
    global youtube  # ensure using fresh auth if credentials switch

    while True:
        try:
            body = {
                "id": video_id,
                "snippet": {
                    "title": title,
                    "description": description,
                    "categoryId": "1"
                }
            }

            request = youtube.videos().update(part="snippet", body=body)
            response = request.execute()
            quota_usage["videos.update"] += 50
            print(f"‚úÖ Updated metadata: {title}")

            # Set Thumbnail if available
            if thumbnail_url:
                try:
                    thumbnail_response = requests.get(thumbnail_url, stream=True)
                    if thumbnail_response.status_code == 200:
                        thumbnail_path = os.path.join(temp_dir, "thumbnail.jpg")
                        with open(thumbnail_path, "wb") as f:
                            for chunk in thumbnail_response.iter_content(1024):
                                f.write(chunk)
                        youtube.thumbnails().set(videoId=video_id, media_body=thumbnail_path).execute()
                        quota_usage["thumbnails.set"] += 50
                        print(f"üñºÔ∏è Thumbnail updated for: {title}")
                        if os.path.exists(thumbnail_path):
                            os.remove(thumbnail_path)
                except Exception as e:
                    print(f"‚ö†Ô∏è Thumbnail update failed: {e}")

            return  # ‚úÖ success
        except HttpError as e:
            if e.resp.status == 403 and "quota" in str(e).lower():
                print("‚õî Quota exceeded during metadata update. Switching credentials...")
                load_valid_credentials(start_index=current_credential_index + 1)
                continue
            else:
                raise

# ============================
# ENHANCED Function to Create Filtered Video in Temp Space
# ============================
def modify_video(video_file):
    # Create a filename for the temporary file
    file_basename = os.path.basename(video_file)
    temp_output = os.path.join(temp_dir, f"filtered_{file_basename}")

    # Extract only the desired audio track
    command = [
        "ffmpeg", "-i", video_file, "-map", "0:v:0", "-map", f"0:a:m:language:{selected_audio_language}",
        "-c:v", "copy", "-c:a", "copy", "-y", temp_output
    ]

    print(f"üîÑ Creating filtered version of {file_basename}...")

    # Get video duration to estimate progress
    try:
        duration_cmd = ["ffprobe", "-v", "error", "-show_entries", "format=duration", "-of", "default=noprint_wrappers=1:nokey=1", video_file]
        duration = float(subprocess.check_output(duration_cmd).decode().strip())

        # Create process
        process = subprocess.Popen(
            command,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            universal_newlines=True
        )

        # Create progress bar
        filter_pbar = tqdm(total=100, desc="Filtering video", colour="yellow")

        # Parse ffmpeg output to update progress bar
        for line in process.stdout:
            # Try to extract time information
            time_match = re.search(r"time=(\d+):(\d+):(\d+)", line)
            if time_match:
                h, m, s = map(int, time_match.groups())
                current_time = h * 3600 + m * 60 + s
                progress = min(int(current_time / duration * 100), 100)
                filter_pbar.update(progress - filter_pbar.n)

        # Close progress bar and ensure process is done
        filter_pbar.close()
        process.wait()

        # Check if output file was created successfully
        if os.path.exists(temp_output):
            print(f"‚úÖ Filtered version created at {temp_output}")
        else:
            print(f"‚ùå Failed to create filtered version")

    except Exception as e:
        print(f"‚ùå Error during video filtering: {e}")
        # Fallback to standard subprocess.run if the progress approach fails
        subprocess.run(command, check=True)

    # Return the path to the temporary filtered file
    return temp_output

# ============================
# ENHANCED Function to Upload Video & Subtitles
# ============================
def upload_video(video_file, subtitles, index):
    global youtube  # ensure we're always using the latest auth

    while True:  # keep retrying until successful or unrecoverable error
        try:
            video_name = get_upload_name(video_file, index)

            body = {
                "snippet": {
                    "title": video_name,
                    "description": "",
                    "tags": [],
                    "categoryId": "1"
                },
                "status": {
                    "privacyStatus": "private"
                }
            }

            media = MediaFileUpload(video_file, chunksize=-1, resumable=True, mimetype="video/*")
            request = youtube.videos().insert(part="snippet,status", body=body, media_body=media)

            upload_pbar = tqdm(total=100, desc=f"Uploading {os.path.basename(video_file)}", colour="red")
            response = None
            while response is None:
                status, response = request.next_chunk()
                if status:
                    progress = int(status.progress() * 100)
                    upload_pbar.update(progress - upload_pbar.n)

            upload_pbar.close()
            print(f"‚úÖ Video Upload Complete: {video_name}")
            video_id = response["id"]
            quota_usage["videos.insert"] += 1600

            # ===== Subtitle Upload (same as before) =====
            if "Upload" in subtitle_option and subtitles:
                print(f"Uploading {len(subtitles)} subtitle(s) for {video_name}...")
                subtitle_upload_pbar = tqdm(subtitles, desc="Uploading subtitles", colour="magenta")

                for subtitle_file in subtitle_upload_pbar:
                    subtitle_basename = os.path.basename(subtitle_file)
                    language_name = os.path.splitext(subtitle_basename)[0]
                    language_code = "en"

                    lang_name_to_code = {v: k for k, v in {
                        "eng": "English", "hin": "Hindi", "spa": "Spanish",
                        "fra": "French", "ger": "German"
                    }.items()}
                    if language_name in lang_name_to_code:
                        language_code = lang_name_to_code[language_name]

                    subtitle_upload_pbar.set_description(f"Uploading {language_name}")

                    caption_body = {
                        "snippet": {
                            "videoId": video_id,
                            "language": language_code,
                            "name": language_name
                        }
                    }

                    try:
                        caption_request = youtube.captions().insert(
                            part="snippet",
                            body=caption_body,
                            media_body=MediaFileUpload(subtitle_file, mimetype="application/octet-stream")
                        )
                        caption_request.execute()
                        quota_usage["captions.insert"] += 50
                        subtitle_upload_pbar.set_description(f"‚úÖ Uploaded {language_name}")
                    except Exception as e:
                        subtitle_upload_pbar.set_description(f"‚ùå Failed {language_name}")
                        print(f"Error uploading subtitle: {e}")

            return video_id, video_name  # ‚¨ÖÔ∏è Success, return here

        except HttpError as e:
            if e.resp.status == 403 and "quota" in str(e).lower():
                print("‚õî Quota exhausted during video upload.")
                print("üîÑ Switching to next available credentials...")
                load_valid_credentials(start_index=current_credential_index + 1)
                continue  # retry with new credentials
            else:
                raise  # re-raise other errors

# ============================
# Process Subtitles First
# ============================
all_subtitles = {}
print("üîç Analyzing and extracting subtitles...")
for video in tqdm(video_files, desc="Processing videos", colour="cyan"):
    print(f"\nProcessing subtitles for {os.path.basename(video)}...")
    subtitles = []
    if subtitle_option == "Upload All Subtitles" or subtitle_option == "Extract Subtitles":
        subtitles = extract_subtitles(video, english_only=False)
    elif subtitle_option == "Upload English Subtitles" or subtitle_option == "Extract English Subtitles":
        subtitles = extract_subtitles(video, english_only=True)
    all_subtitles[video] = subtitles

# ============================
# Fetch Metadata Based on Video Type
# ============================
movie_metadata = None
episodes_metadata = None
is_series = False

if fetch_metadata and tmdb_link:
    print("üîç Fetching metadata from TMDB...")
    # First, determine if we're dealing with series or movie
    if any(is_series_episode(video) for video in video_files):
        is_series = True
        # Create the content link for fetching episode metadata by appending /season/X
        content_tmdb_link = f"{tmdb_link}/season/{season_number}"

        # Parse TV show link
        tv_info = parse_tmdb_tv_link(content_tmdb_link)
        if tv_info:
            print(f"üì∫ Detected TV series, fetching metadata for season {tv_info['season_number']}...")
            episodes_metadata = fetch_episodes(tv_info['id'], tv_info['season_number'], tmdb_api_key)

            if episodes_metadata:
                print(f"‚úÖ Found metadata for {len(episodes_metadata)} episodes")
            else:
                print("‚ö†Ô∏è Could not fetch episode metadata")
    else:
        # Parse movie link - use the link as is for movies
        movie_id = parse_tmdb_movie_link(tmdb_link)
        if movie_id:
            print(f"üé¨ Detected movie, fetching metadata...")
            movie_metadata = fetch_movie_metadata(movie_id, tmdb_api_key)

            if movie_metadata:
                print(f"‚úÖ Found movie metadata: {movie_metadata['title']}")
            else:
                print("‚ö†Ô∏è Could not fetch movie metadata")

# ============================
# Process and Upload Videos
# ============================
uploaded_video_ids = []
temp_files = []  # Track temporary files for cleanup

print(f"\nüîÑ Processing {len(video_files)} video(s)...")
for i, video in enumerate(video_files):
    print(f"\n[{i+1}/{len(video_files)}] Processing {os.path.basename(video)}...")

    # Create modified version in temp space
    temp_video = modify_video(video)
    temp_files.append(temp_video)

    # Upload from temp location
    video_id, video_name = upload_video(temp_video, all_subtitles.get(video, []), i)
    uploaded_video_ids.append(video_id)

    # Update metadata if available
    if fetch_metadata and tmdb_link:
        if is_series and episodes_metadata and video_name.isdigit() and video_name in episodes_metadata:
            # If it's an episode and we have matching metadata
            episode_info = episodes_metadata[video_name]
            print(f"üîÑ Updating metadata for episode {video_name}...")
            update_video_metadata(
                video_id,
                episode_info["title"],
                episode_info["description"],
                episode_info["thumbnail"]
            )
        elif not is_series and movie_metadata:
            # If it's a movie and we have metadata
            print("üîÑ Updating metadata for movie...")
            update_video_metadata(
                video_id,
                movie_metadata["title"],
                movie_metadata["description"],
                movie_metadata["thumbnail"]
            )

# Handle playlist operations after all videos are uploaded

# Sort uploaded videos by their episode number before adding to playlist
if is_series:
    # Create a list of tuples (video_name as int, video_id), then sort
    sorted_videos = sorted(
        zip([int(get_upload_name(video, i)) for i, video in enumerate(video_files)], uploaded_video_ids),
        key=lambda x: x[0]
    )
    sorted_video_ids = [vid for _, vid in sorted_videos]
else:
    sorted_video_ids = uploaded_video_ids

# Add to playlist in sorted order
handle_playlist(playlist_option, sorted_video_ids)


# Clean up temporary files
print("\nüßπ Cleaning up temporary files...")
cleanup_pbar = tqdm(temp_files, desc="Cleaning temporary files", colour="green")
for temp_file in cleanup_pbar:
    try:
        if os.path.exists(temp_file):
            os.remove(temp_file)
            cleanup_pbar.set_description(f"Removed {os.path.basename(temp_file)}")
    except Exception as e:
        cleanup_pbar.set_description(f"Error: {str(e)[:20]}...")

# Clean up subtitle folders
subtitle_dirs = [d for d in os.listdir(temp_dir) if d.startswith("subtitles_")]
cleanup_pbar = tqdm(subtitle_dirs, desc="Cleaning subtitle directories", colour="green")
for subtitle_dir in cleanup_pbar:
    dir_path = os.path.join(temp_dir, subtitle_dir)
    try:
        # Remove all files in the directory
        files_in_dir = os.listdir(dir_path)
        files_pbar = tqdm(files_in_dir, desc=f"Removing files in {subtitle_dir}", leave=False, colour="cyan")
        for file in files_pbar:
            file_path = os.path.join(dir_path, file)
            if os.path.isfile(file_path):
                os.remove(file_path)
                files_pbar.set_description(f"Removed {file}")
        # Remove the directory
        os.rmdir(dir_path)
        cleanup_pbar.set_description(f"Removed directory {subtitle_dir}")
    except Exception as e:
        cleanup_pbar.set_description(f"Error with {subtitle_dir}: {str(e)[:20]}...")

# Try to remove the temp directory
try:
    os.rmdir(temp_dir)
    print(f"‚úÖ Removed temporary directory: {temp_dir}")
except OSError as e:
    print(f"‚ö†Ô∏è Note: Could not remove temp directory: {e}")

# ============================
# Display YouTube API Quota Summary
# ============================
print("\nüìä YOUTUBE API QUOTA USAGE SUMMARY")
print("‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ")

quota_total = 0
quota_limits = {
    "videos.insert": 1600,
    "captions.insert": 50,
    "videos.update": 50,
    "thumbnails.set": 50,
    "playlists.insert": 50,
    "playlistItems.insert": 50
}

for action, count in quota_usage.items():
    cost = quota_limits[action] * (count // quota_limits[action] if quota_limits[action] > 0 else 0)
    quota_total += cost
    print(f"{action:22s}: {count // quota_limits[action]} √ó {quota_limits[action]} = {cost} units")

print("‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ")
print(f"üì¶ Total estimated quota used: {quota_total} / 10,000 units")
if quota_total > 10000:
    print("‚ö†Ô∏è Quota exceeded. Additional credentials were used.")
elif quota_total > 8000:
    print("‚ö†Ô∏è Near quota limit. Consider switching credentials soon.")
else:
    print("‚úÖ Quota usage within safe limits.")
print("‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ")

print("\nüéâ All processing complete!")
```

```python
#@markdown <h1><strong><center> Movies/Series Uploader üìΩÔ∏è
import os
import pickle
import subprocess
from google.auth.transport.requests import Request
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from langdetect import detect

# ============================
# User Input: Video Path, Language Selection & Subtitle Option
# ============================
video_path = ""  #@param {type:"string"}
selected_audio_language = "hin"  #@param ["hin", "eng", "fra", "spa"]
subtitle_option = "Upload English Subtitles"  #@param ["Upload All Subtitles", "Upload English Subtitles", "Extract Subtitles", "Extract English Subtitles", "No Subtitles"]

# Ensure correct directory handling
if os.path.isdir(video_path):
    video_dir = video_path
    video_files = [os.path.join(video_dir, f) for f in os.listdir(video_dir) if f.endswith(('.mp4', '.mkv'))]
elif os.path.isfile(video_path):
    video_dir = os.path.dirname(video_path)
    video_files = [video_path]
else:
    raise FileNotFoundError("Invalid path! Provide a valid video file or directory.")

# ============================
# Authentication & API Setup
# ============================
credentials_number = 0  #@param {type:"integer"}
credentials_path = f"/content/drive/MyDrive/Credentials/credentials{credentials_number}.json"

if os.path.exists("token.pickle"):
    with open("token.pickle", "rb") as token:
        try:
            saved_creds = pickle.load(token)
            if saved_creds and saved_creds.token and saved_creds.client_id:
                with open(credentials_path, "r") as f:
                    if saved_creds.client_id not in f.read():
                        os.remove("token.pickle")
                        print("Credentials changed. Deleted token.pickle.")
        except Exception:
            os.remove("token.pickle")
            print("Error reading token.pickle. Deleted token file.")

SCOPES = ["https://www.googleapis.com/auth/youtube.force-ssl"]
creds = None
if os.path.exists("token.pickle"):
    with open("token.pickle", "rb") as token:
        creds = pickle.load(token)
if not creds or not creds.valid:
    if creds and creds.expired and creds.refresh_token:
        creds.refresh(Request())
    else:
        flow = InstalledAppFlow.from_client_secrets_file(credentials_path, SCOPES)
        creds = flow.run_console()
    with open("token.pickle", "wb") as token:
        pickle.dump(creds, token)

youtube = build("youtube", "v3", credentials=creds)

# ============================
# Function to Extract Subtitles
# ============================
def extract_subtitles(video_file, english_only=False):
    subtitle_folder = os.path.join(os.path.dirname(video_file), f"[Subtitles] - {os.path.basename(video_file)}")
    os.makedirs(subtitle_folder, exist_ok=True)
    extracted_files = []

    if video_file.endswith(".mkv"):
        try:
            track_info = subprocess.check_output(["mkvmerge", "-i", video_file]).decode()
            subtitle_tracks = [line for line in track_info.split("\n") if "subtitles" in line.lower()]

            for i, track in enumerate(subtitle_tracks):
                track_id = track.split(':')[0].split()[-1]
                subtitle_ext = "ass" if "ass" in track.lower() else "srt"
                subtitle_name = f"subtitle_{i}.{subtitle_ext}"
                output_file = os.path.join(subtitle_folder, subtitle_name)

                extract_cmd = ["mkvextract", "tracks", video_file, f"{track_id}:{output_file}"]
                result = subprocess.run(extract_cmd, check=False, capture_output=True, text=True)

                if result.returncode == 0 and os.path.exists(output_file):
                    if english_only and detect(open(output_file, "r", encoding="utf-8", errors="ignore").read()) != "en":
                        os.remove(output_file)
                        continue
                    new_output_file = os.path.join(subtitle_folder, f"[Extracted] - subtitle_{i}.{subtitle_ext}")
                    os.rename(output_file, new_output_file)
                    extracted_files.append(new_output_file)
                    print(f"‚úÖ Subtitle extracted: {new_output_file}")
        except subprocess.CalledProcessError as e:
            print(f"‚ùå MKVToolNix error: {e}")
    return extracted_files

# ============================
# Process Subtitles First
# ============================
for video in video_files:
    print(f"Processing subtitles for {video}...")
    subtitles = []
    if subtitle_option == "Upload All Subtitles" or subtitle_option == "Extract Subtitles":
        subtitles = extract_subtitles(video, english_only=False)
    elif subtitle_option == "Upload English Subtitles" or subtitle_option == "Extract English Subtitles":
        subtitles = extract_subtitles(video, english_only=True)

# ============================
# Function to Modify Video (Keep Only Selected Audio)
# ============================
def modify_video(video_file):
    temp_output = f"{video_file.rsplit('.', 1)[0]}_filtered.mkv"
    command = [
        "ffmpeg", "-i", video_file, "-map", "0:v:0", "-map", f"0:a:m:language:{selected_audio_language}",
        "-c:v", "copy", "-c:a", "copy", "-y", temp_output
    ]
    subprocess.run(command, check=True)
    os.replace(temp_output, video_file)
    print(f"‚úÖ Modified video: {video_file}")

# ============================
# Function to Upload Video & Subtitles
# ============================
def upload_video(video_file, subtitles):
    video_name = os.path.splitext(os.path.basename(video_file))[0]
    body = {"snippet": {"title": video_name, "description": "", "tags": [], "categoryId": "1"},
        "status": {"privacyStatus": "private"}}
    media = MediaFileUpload(video_file, chunksize=-1, resumable=True, mimetype="video/*")
    request = youtube.videos().insert(part="snippet,status", body=body, media_body=media)
    response = None
    while response is None:
        status, response = request.next_chunk()
        if status:
            print(f"Uploaded {int(status.progress() * 100)}%")
    print("‚úÖ Video Upload Complete")
    video_id = response["id"]

    if "Upload" in subtitle_option:
        for subtitle_file in subtitles:
            caption_body = {"snippet": {"videoId": video_id, "language": "en", "name": os.path.basename(subtitle_file)}}
            caption_request = youtube.captions().insert(part="snippet", body=caption_body, media_body=MediaFileUpload(subtitle_file, mimetype="application/octet-stream"))
            caption_request.execute()
            print(f"‚úÖ Subtitle '{os.path.basename(subtitle_file)}' uploaded.")

# ============================
# Process and Upload Videos
# ============================
for video in video_files:
    print(f"Processing {video}...")
    modify_video(video)
    upload_video(video, subtitles)
```

```python
#@markdown <h1> <center> <strong> Auto Renamer üìù
import os
import re

# Function to rename files based on the episode number (EXX)
def rename_mkv_files(folder_path: str):
    """
    Renames .mkv files in the specified folder based on their episode number (EXX) in the SXXEXX format.

    @param folder_path: The path to the folder containing the .mkv files.
    """
    if not os.path.exists(folder_path):
        print("Folder does not exist!")
        return

    # Regex pattern to match filenames with SXXEXX format
    pattern = re.compile(r"S\d{2}E(\d{2})\.mkv", re.IGNORECASE)

    for filename in os.listdir(folder_path):
        match = pattern.search(filename)
        if match:
            episode_number = match.group(1)  # Extract episode number (XX)
            new_filename = f"{int(episode_number)}.mkv"  # Convert to int to remove leading zeros
            old_path = os.path.join(folder_path, filename)
            new_path = os.path.join(folder_path, new_filename)

            if not os.path.exists(new_path):  # Avoid overwriting existing files
                os.rename(old_path, new_path)
                print(f"Renamed: {filename} -> {new_filename}")
            else:
                print(f"Skipping: {new_filename} already exists!")
        else:
            print(f"Skipping: {filename} (does not match pattern or wrong extension)")

# Get user input for folder path in Google Colab
folder_path = ""  #@param {type:"string"}
rename_mkv_files(folder_path)
```

```python
#@markdown <h1> <strong> <center> Subtitles Extractor

# ==========================
# üîß Required Installation
# ==========================
!apt-get install -y mkvtoolnix mkvtoolnix-gui
!pip install tqdm

import os
import re
import subprocess
from collections import defaultdict
from tqdm.notebook import tqdm

# ========== üì• FORM INPUTS ==========
input_path = ""  # @param {type: "string"}
save_path = ""  # @param {type: "string"}
extract_option = "English"  # @param ["All", "English"]

# ========== üîß MAIN FUNCTION ==========
def extract_subtitles(mkv_file, extract_option, output_dir=None):
    filename = os.path.basename(mkv_file)
    file_base = os.path.splitext(filename)[0]

    # Detect subtitle tracks
    result = subprocess.run(['mkvmerge', '-i', mkv_file], capture_output=True, text=True)
    lines = result.stdout.strip().splitlines()

    subtitle_tracks = []
    for line in lines:
        if 'subtitles' in line.lower():
            lang = "Unknown"
            lang_match = re.search(r'\[language: (.*?)\]', line.lower())
            if lang_match:
                lang = lang_match.group(1).capitalize()

            if extract_option == 'English':
                if lang.lower() in ["en", "eng", "english", "unknown"]:
                    subtitle_tracks.append((line, lang))
            else:
                subtitle_tracks.append((line, lang))

    if not subtitle_tracks:
        print(f"‚ö†Ô∏è No matching subtitles found in {filename}")
        return

    # Group by language
    lang_counts = defaultdict(int)

    for line, lang in subtitle_tracks:
        track_id_match = re.search(r'Track ID (\d+):', line)
        if not track_id_match:
            continue

        track_id = track_id_match.group(1)
        lang = lang.capitalize()
        lang_counts[lang] += 1
        lang_index = lang_counts[lang]

        folder_name = f"[Subtitle] - {lang} {file_base}"
        output_folder = os.path.join(output_dir or os.path.dirname(mkv_file), folder_name)
        os.makedirs(output_folder, exist_ok=True)

        # If only one subtitle for this language, don't add number
        display_lang = f"{lang} {lang_index}" if lang_counts[lang] > 1 else lang
        output_file = os.path.join(output_folder, f"{file_base}.{display_lang}.srt")

        subprocess.run([
            'mkvextract', 'tracks', mkv_file,
            f"{track_id}:{output_file}"
        ], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# ========== üöÄ RUN EXTRACTION ==========
input_path = input_path.strip()
save_path = save_path.strip() or None

if not os.path.exists(input_path):
    print("‚ùå Error: Path doesn't exist.")
else:
    mkv_files = []
    if os.path.isfile(input_path) and input_path.endswith('.mkv'):
        mkv_files = [input_path]
    elif os.path.isdir(input_path):
        for root, _, files in os.walk(input_path):
            for file in files:
                if file.endswith('.mkv'):
                    mkv_files.append(os.path.join(root, file))

    if not mkv_files:
        print("‚ùå No MKV files found.")
    else:
        print(f"üîç Found {len(mkv_files)} MKV file(s). Starting...")
        for mkv in tqdm(mkv_files, desc="Extracting Subtitles"):
            extract_subtitles(mkv, extract_option, save_path)
        print("‚úÖ Done! Subtitles extracted.")
```

```python
# ============================= FORM ============================= #
# @markdown #### ‚¨ÖÔ∏è Extract Files
MODE = "UNZIP"  # @param ["UNZIP", "UNTAR", "UNRAR", "7Z"]
PATH_TO_FILE = ""  # @param {type:"string"}
extractPath = ""  # @param {type:"string"}
ARCHIVE_PASSWORD = "" #@param {type:"string"}
DELETE_ORIGINAL = True #@param {type:"boolean"}

# ================================================================ #
import os
import urllib.request
import shutil
import tempfile
import time
import sys
import threading
import subprocess
from google.colab import drive
import re

HOME = os.path.expanduser("~")

# Install tqdm if not already installed
try:
    from tqdm.notebook import tqdm  # Use notebook version for better visuals
except ImportError:
    subprocess.run([sys.executable, "-m", "pip", "install", "tqdm"], check=True)
    from tqdm.notebook import tqdm  # Import notebook version after installation

# Setting up helper functions
if not os.path.exists(f"{HOME}/.ipython/ocr.py"):
    hCode = "https://raw.githubusercontent.com/biplobsd/" \
                "OneClickRun/master/res/ocr.py"
    urllib.request.urlretrieve(hCode, f"{HOME}/.ipython/ocr.py")

from ocr import checkAvailable

# Modified runSh function with improved progress bar support
def runSh(cmd, output=True):
    """Run shell commands with a progress bar for specific operations"""
    # Commands that involve extraction
    extraction_commands = ['unzip', 'unrar', '7z', 'tar']

    # Check if this is an extraction command
    is_extraction = any(cmd.startswith(ex_cmd) for ex_cmd in extraction_commands)

    if is_extraction and output:
        # Create a process with pipe for output
        process = subprocess.Popen(
            cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT,
            universal_newlines=True, bufsize=1
        )

        # Initialize progress variables
        total_files = 0
        processed_files = 0

        # First pass to estimate total files (if possible)
        if cmd.startswith('unzip -l'):
            # For unzip, we can get total files from list command
            list_cmd = f'unzip -l "{PATH_TO_FILE}" | wc -l'
            try:
                total_files = int(subprocess.check_output(list_cmd, shell=True).decode().strip()) - 4  # Adjust for header/footer
                if total_files < 0:
                    total_files = 0
            except:
                total_files = 0

        # Create a progress bar
        if total_files > 0:
            pbar = tqdm(total=total_files, desc="Extracting", unit='files', colour='green',
                        bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}]')
        else:
            # If we can't determine file count, use a spinner-style progress
            pbar = tqdm(total=100, desc="Extracting", unit='%', colour='blue',
                        bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt}% [{elapsed}<{remaining}]')
            # Start a thread to increment the bar periodically
            def update_spinner():
                progress = 0
                while progress < 99 and process.poll() is None:
                    time.sleep(0.5)
                    progress += 1
                    pbar.update(1)

            spinner_thread = threading.Thread(target=update_spinner)
            spinner_thread.daemon = True
            spinner_thread.start()

        # Read and process output
        for line in process.stdout:
            # Update progress based on output
            if cmd.startswith('unzip'):
                if line.startswith('  inflating:') or line.startswith(' extracting:'):
                    processed_files += 1
                    if total_files > 0:
                        pbar.update(1)
            elif cmd.startswith('unrar'):
                if 'Extracting' in line:
                    processed_files += 1
                    if total_files > 0:
                        pbar.update(1)
                    else:
                        pbar.update(0)  # Just to refresh
            elif cmd.startswith('7z'):
                if 'Extracting' in line:
                    processed_files += 1
                    if total_files > 0:
                        pbar.update(1)
                    else:
                        pbar.update(0)  # Just to refresh
            elif cmd.startswith('tar'):
                processed_files += 1
                if total_files > 0:
                    pbar.update(1)
                else:
                    pbar.update(0)  # Just to refresh

        process.wait()
        pbar.close()

        if total_files == 0:
            print(f"‚úÖ Extraction completed. Processed approximately {processed_files} files.")
        else:
            print(f"‚úÖ Extraction completed. Processed {processed_files}/{total_files} files.")

        returncode = process.returncode
        if returncode != 0:
            raise subprocess.CalledProcessError(returncode, cmd)

        return returncode
    else:
        # For non-extraction commands, use the regular approach
        process = subprocess.run(cmd, shell=True, check=True, text=True, stdout=subprocess.PIPE if output else None)
        if output:
            return process.stdout
        return process.returncode

# Function to get file size
def get_file_size(file_path):
    """Get the size of a file in bytes"""
    try:
        return os.path.getsize(file_path)
    except:
        return 0

# Function to permanently delete a file from Google Drive with progress bar
def permanently_delete_file(file_path):
    """
    Permanently deletes a file from Google Drive without moving to trash with progress bar
    """
    file_size = get_file_size(file_path)

    try:
        # Check if the file is in Google Drive path
        if file_path.startswith('/content/drive/'):
            print(f"üóëÔ∏è Preparing to delete: {file_path}")

            # Create a progress bar for deletion
            with tqdm(total=100, desc="Deleting original file", unit="%", colour='red',
                     bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt}% [{elapsed}<{remaining}]') as pbar:
                # Update to 10% to show deletion has started
                pbar.update(10)

                # Ensure drive is mounted
                if not os.path.exists('/content/drive'):
                    drive.mount('/content/drive')

                # Update to 20% after drive mount check
                pbar.update(10)

                # Use Google Drive API to permanently delete the file
                from googleapiclient.discovery import build
                from google.colab import auth
                from googleapiclient.errors import HttpError

                # Authenticate
                auth.authenticate_user()
                pbar.update(20)  # 40% progress

                # Build the Drive service
                drive_service = build('drive', 'v3')
                pbar.update(10)  # 50% progress

                # Get the file ID from the path using the Drive API
                # Search for the file by name in Google Drive
                file_name = os.path.basename(file_path)
                pbar.update(10)  # 60% progress

                response = drive_service.files().list(
                    q=f"name='{file_name}' and trashed=false",
                    spaces='drive',
                    fields='files(id, name, parents)'
                ).execute()

                files = response.get('files', [])
                file_id = None
                pbar.update(10)  # 70% progress

                # If multiple files with the same name exist, try to find the best match
                if files:
                    # If there's only one file, use it
                    if len(files) == 1:
                        file_id = files[0]['id']
                    else:
                        # Try to match by full path components
                        path_parts = os.path.normpath(file_path).split(os.sep)
                        path_parts = [p for p in path_parts if p and p != '.' and p != '..']

                        for file in files:
                            # Get file path information to compare
                            if file_name == file['name']:
                                file_id = file['id']
                                break

                pbar.update(10)  # 80% progress

                if file_id:
                    drive_service.files().delete(fileId=file_id, supportsAllDrives=True).execute()
                    pbar.update(20)  # 100% progress
                    print(f"‚úÖ File permanently deleted: {file_path} ({file_size / (1024*1024):.2f} MB)")
                else:
                    pbar.update(10)  # 90% progress
                    print(f"‚ö†Ô∏è Could not find Google Drive ID for: {file_path}")
                    # Fallback to regular deletion if ID not found
                    os.remove(file_path)
                    pbar.update(10)  # 100% progress
        else:
            # For non-Google Drive files, use regular deletion with progress
            with tqdm(total=100, desc="Deleting original file", unit="%", colour='red',
                     bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt}% [{elapsed}<{remaining}]') as pbar:
                pbar.update(50)  # 50% progress immediately
                os.remove(file_path)
                pbar.update(50)  # Complete the progress
                print(f"‚úÖ File deleted: {file_path} ({file_size / (1024*1024):.2f} MB)")
    except Exception as e:
        print(f"‚ùå Error deleting file {file_path}: {str(e)}")
        # Fallback to regular deletion
        try:
            print("‚ö†Ô∏è Attempting fallback deletion method...")
            with tqdm(total=100, desc="Fallback deletion", unit="%", colour='yellow',
                     bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt}% [{elapsed}<{remaining}]') as pbar:
                pbar.update(50)
                os.remove(file_path)
                pbar.update(50)
                print(f"‚úÖ File removed using fallback method: {file_path}")
        except:
            print(f"‚ùå Failed to delete file even with fallback method: {file_path}")

# Function to move files with progress tracking
def move_files(source_dir, dest_dir):
    """Move files from source to destination with progress bar"""
    # Get a list of all items to move
    items = os.listdir(source_dir)
    total_items = len(items)

    if total_items == 0:
        print(f"‚ÑπÔ∏è No items to move from {source_dir}")
        return

    # Calculate total size for better progress tracking
    total_size = 0
    for item in items:
        item_path = os.path.join(source_dir, item)
        if os.path.isfile(item_path):
            total_size += os.path.getsize(item_path)
        else:
            # For directories, count all files recursively
            for dirpath, _, filenames in os.walk(item_path):
                for f in filenames:
                    total_size += os.path.getsize(os.path.join(dirpath, f))

    # Use size-based progress if we have a reasonable size
    if total_size > 0:
        with tqdm(total=total_size, unit='B', unit_scale=True, desc="Moving files", colour='cyan',
                 bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}]') as pbar:
            for item in items:
                source = os.path.join(source_dir, item)
                destination = os.path.join(dest_dir, item)

                # Handle case where destination already exists
                if os.path.exists(destination):
                    if os.path.isdir(source) and os.path.isdir(destination):
                        # Move directory contents
                        for subitem in os.listdir(source):
                            subitem_source = os.path.join(source, subitem)
                            size = get_dir_size(subitem_source) if os.path.isdir(subitem_source) else os.path.getsize(subitem_source)
                            shutil.move(subitem_source, destination)
                            pbar.update(size)
                    else:
                        # Replace file
                        size = os.path.getsize(source)
                        shutil.move(source, destination)
                        pbar.update(size)
                else:
                    # Move item directly
                    size = get_dir_size(source) if os.path.isdir(source) else os.path.getsize(source)
                    shutil.move(source, destination)
                    pbar.update(size)
    else:
        # Fall back to item count based progress
        with tqdm(total=total_items, unit='items', desc="Moving files", colour='cyan',
                 bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} items [{elapsed}<{remaining}]') as pbar:
            for item in items:
                source = os.path.join(source_dir, item)
                destination = os.path.join(dest_dir, item)

                # Handle case where destination already exists
                if os.path.exists(destination):
                    if os.path.isdir(source) and os.path.isdir(destination):
                        # Move directory contents
                        for subitem in os.listdir(source):
                            shutil.move(os.path.join(source, subitem), destination)
                    else:
                        # Replace file
                        shutil.move(source, destination)
                else:
                    # Move item directly
                    shutil.move(source, destination)

                pbar.update(1)

    print(f"‚úÖ All files moved to final extraction path: {dest_dir}")

# Helper function to get directory size
def get_dir_size(path):
    """Calculate total size of a directory recursively"""
    total_size = 0
    for dirpath, _, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            if os.path.exists(fp):  # Check to avoid broken symlinks
                total_size += os.path.getsize(fp)
    return total_size

def extractFiles():
    global extractPath

    # Ensure the paths are set properly
    if not PATH_TO_FILE:
        raise ValueError("‚ùå Please provide the path to the archive file")

    if not os.path.exists(PATH_TO_FILE):
        raise FileNotFoundError(f"‚ùå Archive file not found: {PATH_TO_FILE}")

    # Get archive file size for information
    archive_size = get_file_size(PATH_TO_FILE)
    print(f"üì¶ Archive size: {archive_size / (1024*1024):.2f} MB")

    # Set default extraction path if not provided
    if not extractPath:
        extractPath = "/content/extract"

    # Use the boolean value directly
    should_delete = DELETE_ORIGINAL

    # If we're not deleting the original, extract directly to the final path
    if not should_delete:
        os.makedirs(extractPath, exist_ok=True)

        # Prepare password parameter if provided
        if ARCHIVE_PASSWORD:
            passADD = f'-P {ARCHIVE_PASSWORD}'
        else:
            passADD = ''

        try:
            # Extract directly to the final path
            if MODE == "UNZIP":
                print(f"üîÑ Extracting ZIP archive to {extractPath}...")
                runSh(f'unzip {passADD} "{PATH_TO_FILE}" -d "{extractPath}"', output=True)
            elif MODE == "UNRAR":
                print(f"üîÑ Extracting RAR archive to {extractPath}...")
                runSh(f'unrar x "{PATH_TO_FILE}" "{extractPath}" {passADD} -o+', output=True)
            elif MODE == "UNTAR":
                print(f"üîÑ Extracting TAR archive to {extractPath}...")
                runSh(f'tar -C "{extractPath}" -xvf "{PATH_TO_FILE}"', output=True)
            else:  # 7Z is the default
                print(f"üîÑ Extracting 7Z archive to {extractPath}...")
                runSh(f'7z x "{PATH_TO_FILE}" -o{extractPath} {passADD}', output=True)

            print(f"‚úÖ Extraction completed to: {extractPath}")

        except Exception as e:
            print(f"‚ùå Error during extraction: {str(e)}")
            raise
    else:
        # Create a temporary directory for extraction (original behavior when deleting)
        temp_dir = tempfile.mkdtemp(dir="/content")
        print(f"üìÇ Created temporary directory for extraction: {temp_dir}")

        # Prepare password parameter if provided
        if ARCHIVE_PASSWORD:
            passADD = f'-P {ARCHIVE_PASSWORD}'
        else:
            passADD = ''

        # Create the final extraction directory
        os.makedirs(extractPath, exist_ok=True)

        try:
            # Extract to temporary directory first
            if MODE == "UNZIP":
                print(f"üîÑ Extracting ZIP archive to temporary location...")
                runSh(f'unzip {passADD} "{PATH_TO_FILE}" -d "{temp_dir}"', output=True)
            elif MODE == "UNRAR":
                print(f"üîÑ Extracting RAR archive to temporary location...")
                runSh(f'unrar x "{PATH_TO_FILE}" "{temp_dir}" {passADD} -o+', output=True)
            elif MODE == "UNTAR":
                print(f"üîÑ Extracting TAR archive to temporary location...")
                runSh(f'tar -C "{temp_dir}" -xvf "{PATH_TO_FILE}"', output=True)
            else:  # 7Z is the default
                print(f"üîÑ Extracting 7Z archive to temporary location...")
                runSh(f'7z x "{PATH_TO_FILE}" -o{temp_dir} {passADD}', output=True)

            print(f"‚úÖ Extraction completed to temporary directory: {temp_dir}")

            # Permanently delete the original archive file
            print(f"üóëÔ∏è Preparing to delete original archive: {PATH_TO_FILE}")
            permanently_delete_file(PATH_TO_FILE)

            # Move files from temporary directory to final extraction path with progress
            move_files(temp_dir, extractPath)

        except Exception as e:
            print(f"‚ùå Error during extraction: {str(e)}")
            raise
        finally:
            # Clean up the temporary directory with progress
            try:
                print(f"üßπ Cleaning up temporary directory: {temp_dir}")
                with tqdm(total=100, desc="Cleaning temp dir", unit="%", colour='yellow',
                         bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt}% [{elapsed}<{remaining}]') as pbar:
                    pbar.update(50)
                    shutil.rmtree(temp_dir)
                    pbar.update(50)
                print(f"‚úÖ Temporary directory removed: {temp_dir}")
            except Exception as e:
                print(f"‚ùå Error removing temporary directory: {str(e)}")

# Call the function to extract files
try:
    print(f"üöÄ Starting extraction of {PATH_TO_FILE} using {MODE} mode")
    print(f"üóëÔ∏è Delete original file after extraction: {DELETE_ORIGINAL}")

    # Display a starting progress banner
    print("\n" + "="*60)
    print(f" üöÄ EXTRACTION PROCESS STARTED ".center(60, "="))
    print("="*60 + "\n")

    start_time = time.time()
    extractFiles()
    total_time = time.time() - start_time

    # Display completion banner with timing
    print("\n" + "="*60)
    print(f" ‚úÖ EXTRACTION COMPLETED SUCCESSFULLY ".center(60, "="))
    print(f" ‚è±Ô∏è Total time: {total_time:.2f} seconds ".center(60, "="))
    print("="*60 + "\n")

except Exception as e:
    print("\n" + "="*60)
    print(f" ‚ùå EXTRACTION FAILED ".center(60, "="))
    print("="*60)
    print(f"Failed to complete extraction: {str(e)}")
```

```python
#@title
#@markdown <br><center><img alt="" src="" height="200" alt="jellyfin"/></center>
#@markdown <center><h3>Jellyfin<br />The Free Software Media System.</h3></center><br>
import os, zipfile, tarfile, requests, re
import urllib.request
from IPython.display import clear_output, display
from ipywidgets import Dropdown, Select
from subprocess import Popen
import xml.etree.ElementTree as ET


#####################################
USE_FREE_TOKEN = True  # @param {type:"boolean"}
TOKEN = ""  # @param {type:"string"}
LATEST_VERSION = False  # @param {type:"boolean"}
CUSTOM_VERSION = "10.8.0"  # @param {type:"string"}
#@markdown <small>&emsp;Version are available on https://repo.jellyfin.org/releases/server/linux/versions/stable/combined/</small>
REGION = "SA" #@param ["US", "EU", "AP", "AU", "SA", "JP", "IN"]
PORT_FORWARD = "argotunnel" #@param ["ngrok", "localhost", "argotunnel"]
#@markdown Default <br>&emsp;username : admin

HOME = os.path.expanduser("~")

if not os.path.exists(f"{HOME}/.ipython/ocr.py"):
    hCode = "https://raw.githubusercontent.com/biplobsd/" \
                "OneClickRun/master/res/ocr.py"
    urllib.request.urlretrieve(hCode, f"{HOME}/.ipython/ocr.py")


from ocr import (
    runSh,
    loadingAn,
    PortForward_wrapper,
    findProcess,
    CWD,
    textAn
)

loadingAn()

os.makedirs('tools/jellyfin/temp', exist_ok=True)
os.makedirs('tools/jellyfin/transcodes', exist_ok=True)
os.makedirs('downloads', exist_ok=True)

def ff_loader(name):
  import requests

  url = requests.get(
    "https://ffbinaries.com/api/v1/version/latest"
    ).json()['bin']['linux-64'][name]
  urllib.request.urlretrieve(url, f'tools/jellyfin/L{name}.zip')

  with zipfile.ZipFile(f'tools/jellyfin/L{name}.zip', 'r') as zip_ref:
      zip_ref.extractall('tools/jellyfin/')
  os.remove(f"tools/jellyfin/L{name}.zip")
  runSh("rm -rf tools/jellyfin/__MACOSX")
  os.replace(f"tools/jellyfin/{name}", f"/usr/bin/{name}")
  os.chmod(f"/usr/bin/{name}", 0o755)

def customVersionSelect():
  allVersionPage = requests.get("https://repo.jellyfin.org/releases/server/linux/versions/stable/combined/").text
  allVersionTags = re.findall(r'">(\d+\.\d+\.\d+)', allVersionPage, re.M)
  option_list = sorted(allVersionTags, reverse=True)
  dropdown = Select(description="Choose one:", options=option_list)
  dropdown.observe(main, names='value')
  display(dropdown)

def startService():
  # Starting jellyfin
  if not findProcess("jellyfin", "--ffmpeg"):
    CM = r"tools/jellyfin/jellyfin/jellyfin --service " \
    f"-C tools/jellyfin -d {CWD}/tools/jellyfin"
    Popen(CM.split(), shell=True)


  # START_SERVER
  # Ngrok region 'us','eu','ap','au','sa','jp','in'
  Server = PortForward_wrapper(
      PORT_FORWARD, TOKEN, USE_FREE_TOKEN, [['jellyfin', 8096, 'http']],
      REGION.lower(), [f"{HOME}/.ngrok2/jellyfin.yml", 7097]
  ).start('jellyfin', displayB=True)

def main(version=None):

  # JellyfinMain
  if LATEST_VERSION:
    serverUrl = "https://repo.jellyfin.org/releases/server/linux/stable/combined/"
  elif CUSTOM_VERSION:
    serverUrl = f"https://repo.jellyfin.org/releases/server/linux/versions/stable/combined/{CUSTOM_VERSION}/"
  else:
    serverUrl = f"https://repo.jellyfin.org/releases/server/linux/versions/stable/combined/{version.new}/"

  server = requests.get(serverUrl).text
  server_file = re.search(r'f=\"(.+)\">jellyfin_(.+)_([^-]+)\.tar\.gz<',
                              server, re.MULTILINE)
  server_file_url = server_file.group(1)
  versionTag = server_file.group(2)

  clear_output(True)
  loadingAn()
  textAn(f"Installing v{versionTag} ...", ty='twg')

  outputFile = 'tools/jellyfin/temp/jellyfin.tar.gz'
  urllib.request.urlretrieve(serverUrl+server_file_url, outputFile)
  with tarfile.open(outputFile, 'r:gz') as tar_ref:
      tar_ref.extractall('tools/jellyfin')
  os.renames(f'tools/jellyfin/jellyfin_{versionTag}',
            'tools/jellyfin/jellyfin')
  os.remove(outputFile)

  # Default config data
  defaultConfig = "https://raw.githubusercontent.com/biplobsd" \
  "/OneClickRun/master/res/appdata/jellyfin/data.tar.gz"
  outputFile = 'tools/jellyfin/temp/data.tar.gz'
  urllib.request.urlretrieve(defaultConfig, outputFile)
  with tarfile.open(outputFile, 'r:gz') as tar_ref:
      tar_ref.extractall('tools/jellyfin')
  os.remove(outputFile)

  # https://github.com/rokibulislaam/colab-ffmpeg-cuda
  ## with GPU
  if os.system("nvidia-smi --query-gpu=name --format=csv,noheader")==0:
    encoConfigP = 'tools/jellyfin/config/encoding.xml'
    mytree = ET.parse(encoConfigP)
    myroot = mytree.getroot()

    def addnew(name, value):
      ET.SubElement(myroot, name).text = value

    addnew('TranscodingTempPath', '/content/tools/jellyfin/transcodes')
    addnew('HardwareAccelerationType', 'nvenc')
    addnew('DeinterlaceMethod', 'yadif')
    addnew('EnableDecodingColorDepth10Hevc', 'true')
    addnew('EnableDecodingColorDepth10Vp9', 'true')
    addnew('EnableDecodingColorDepth10Vp9', 'true')
    addnew('EncoderPreset', 'slower')
    addnew('EncodingThreadCount', '0')

    for tagname in myroot.iter('HardwareDecodingCodecs'):
      ET.SubElement(tagname, 'string').text = 'hevc'
      ET.SubElement(tagname, 'string').text = 'mpeg2video'
      ET.SubElement(tagname, 'string').text = 'mpeg4'
      ET.SubElement(tagname, 'string').text = 'vp8'
      ET.SubElement(tagname, 'string').text = 'vp9'

    mytree.write(encoConfigP)

    runSh(
        "git clone --depth 1 --branch main https://github.com/rokibulislaam/colab-ffmpeg-cuda.git tools/colab-ffmpeg-cuda")
    runSh("cp -r tools/colab-ffmpeg-cuda/bin/. /usr/bin/")
    ffmpegPath = '/usr/bin/ffmpeg'
  else:
    ## without GPU
    # ffmpeg
    ff_loader('ffmpeg')

    # ffprobe
    ff_loader('ffprobe')

    ffmpegPath = '/usr/bin/ffmpeg'

  startService()

if not os.path.exists("tools/jellyfin/jellyfin/jellyfin"):
  if LATEST_VERSION or CUSTOM_VERSION:
    main()
  else:
    customVersionSelect()
else:
  startService()
```

```python
URL = "" #@param {type:"string"}
#@markdown <center><h5>Default output path /content/downloads/</h5></center>
OUTPUT_PATH = "" #@param {type:"string"}

import pathlib
import shutil
import hashlib
import requests
from urllib.parse import urlparse
from os import path, mkdir
if not path.exists("/root/.ipython/ocr.py"):
    from subprocess import run
    from shlex import split

    shellCmd = "wget -qq https://raw.githubusercontent.com/biplobsd/" \
                "OneClickRun/master/res/ocr.py \
                    -O /root/.ipython/ocr.py"
    run(split(shellCmd))
from ocr import runSh

def youtubedlInstall():
  if not path.isfile("/usr/local/bin/youtube-dl"):
    cmdC = "rm -rf /content/sample_data/ && " \
            " mkdir -p -m 666 /root/.YouTube-DL/ &&" \
            " apt-get install atomicparsley &&" \
            " curl -L https://yt-dl.org/downloads/latest/youtube-dl " \
            "-o /usr/local/bin/youtube-dl &&" \
            " chmod a+rx /usr/local/bin/youtube-dl"
    get_ipython().system_raw(cmdC)

def aria2Install():
  runSh('apt install -y aria2')

def istmd(URL):
  link = urlparse(URL)

  #YandexDisk
  if link.netloc == "yadi.sk":
    API_ENDPOINT = 'https://cloud-api.yandex.net/v1/disk/public/resources/' \
                    '?public_key={}&path=/{}&offset={}'
    dry = False
    def md5sum(file_path):
        md5 = hashlib.md5()
        with open(file_path, 'rb') as f:
            for chunk in iter(lambda: f.read(128 * md5.block_size), b''):
                md5.update(chunk)
        return md5.hexdigest()


    def check_and_download_file(target_path, url, size, checksum):
        if path.isfile(target_path):
            if size == path.getsize(target_path):
                if checksum == md5sum(target_path):
                    print('URL {}'.format(url))
                    print('skipping correct {}'.format(target_path))
                    return
        if not dry:
            print('URL {}'.format(url))
            print('downloading {}'.format(target_path))
            runSh(f'aria2c -x 16 -s 16 -k 1M -d {OUTPUT_PATH} {url}', output=True)
            # r = requests.get(url, stream=True)
            # with open(target_path, 'wb') as f:
            #     shutil.copyfileobj(r.raw, f)


    def download_path(target_path, public_key, source_path, offset=0):
        print('getting "{}" at offset {}'.format(source_path, offset))
        current_path = path.join(target_path, source_path)
        pathlib.Path(current_path).mkdir(parents=True, exist_ok=True)
        jsn = requests.get(API_ENDPOINT.format(public_key, source_path, offset)).json()
        def try_as_file(j):
            if 'file' in j:
                file_save_path = path.join(current_path, j['name'])
                check_and_download_file(file_save_path, j['file'], j['size'], j['md5'])
                return True
            return False

        # first try to treat the actual json as a single file description
        if try_as_file(jsn):
            return

        # otherwise treat it as a directory
        emb = jsn['_embedded']
        items = emb['items']
        for i in items:
            # each item can be a file...
            if try_as_file(i):
                continue
            # ... or a directory
            else:
                subdir_path = path.join(source_path, i['name'])
                download_path(target_path, public_key, subdir_path)

        # check if current directory has more items
        last = offset + emb['limit']
        if last < emb['total']:
            download_path(target_path, public_key, source_path, last)
    download_path(OUTPUT_PATH, URL, '')
    return False
  return URL

if not OUTPUT_PATH:
  OUTPUT_PATH = "/content/downloads/"

if not URL == "":
  aria2Install()
  youtubedlInstall()
  try:
    mkdir("downloads")
  except FileExistsError:
    pass
  url = istmd(URL)
  if url != False:
    print('URL {}'.format(URL))
    cmdC = f'youtube-dl -o "{OUTPUT_PATH}/%(title)s" {URL} ' \
            '--external-downloader aria2c ' \
            '--external-downloader-args "-x 16 -s 16 -k 1M"'
    runSh(cmdC, output=True)
else:
  print("Please input url")
```
